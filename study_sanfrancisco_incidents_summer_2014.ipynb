# Sanfrancisco incidents data in summer 2014

## I am gonna use below data

```scala
case class Insident(
  IncidntNum: Long,
  Category: String,
  Descript: String,
  DayOfWeek: String,
  Date: String,
  Time: String,
  PdDistrict: String,
  Resolution: String,
  Address: String,
  X: Double,
  Y: Double,
  Location: String,
  PdId: Long
)
val reader = CSVReader.open(new File("./notebooks/study/sanfrancisco_incidents_summer_2014.csv"))
val insidentsList = reader.all().tail.map{ values => 
  Insident(
    values(0).toLong,
    values(1),
    values(2),
    values(3),
    values(4),
    values(5),
    values(6),
    values(7),
    values(8),
    values(9).toDouble,
    values(10).toDouble,
    values(11),
    values(12).toLong
  )
}
reader.close()
insidentsList
```

## Below barchart shows how many insidens happen in each of time. Y axis is time and the values are 0 - 23. X axis is number of insidens. Basically number of insidents has corelation with human activity. In early morning such as 3-6am insidents don't happen so much than noon or evening. 

```scala
val seriesByTimeAndInsidentType = insidentsList.groupBy(in => in.Category).values.map{ list =>
  val category = list.head.Category
  val dataListMap = list.groupBy(in => in.Time.split(":")(0).toInt)
  val dataList = (0 until 24).map{ i =>
    Data(i, dataListMap.get(i).map(_.size).getOrElse(0))
  }
  Series(dataList, chart = "bar", name = Some(category))
}

PlotH(
  Highchart(seriesByTimeAndInsidentType, chart = Chart(zoomType = Zoom.xy), yAxis = None).stack(Stacking.normal),
  sizes=(1000, 800)
)




```

## Below barchart shows how many rate of each insidents in each times. One thing we can see is ratio of "ASSULT" is high in 2-5am.

```scala
PlotH(
  Highchart(seriesByTimeAndInsidentType, chart = Chart(zoomType = Zoom.xy), yAxis = None).stack(Stacking.percent),
  sizes=(1000, 800)
)


```
